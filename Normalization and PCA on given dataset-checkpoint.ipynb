# Removing the outliers using IQR method:

IQR or Inter Quartile Range is the difference between the Q3 (third quarter or 75%) and Q1(first quarter or 25%) of the values in the dataset.This will help remove the outliers from the features of the data set.

```python
import scipy.stats as stats
Q1 = df_var_imput.quantile(q=.25)
Q3 = df_var_imput.quantile(q=.75)
IQR = df_var_imput.apply(stats.iqr)
df_clean = df_var_imput[~((df_var_imput< (Q1-1.5*IQR)) | (df_var_imput > (Q3+1.5*IQR))).any(axis=1)]
#In order to find how many rows are left in the dataframe 
df_clean.shape
```

# Normalization of the Features

After the removal of outliers, data normalization helps in scaling the features in the data in one range of values, for example 0-1. This helps scale up or down the features to bring them on the same level for further data manipulation.

```python
df_norm = (df_clean-df_clean.min())/ (df_clean.max() - df_clean.min())
df_norm.info
df_norm
```

# Principal Component Analysis

PCA is used to identify patterns in a dataset.The goal is to find the components of variance in data. It can be used to reduce the dimensionality of a dataset while preserving information. Also, it is used for data visualization and as a preprocessing step for machine learning.
The first component explains the most variation, and each subsequent component explains less and less variation.

```python
pca = PCA(n_components=0.7)
scores = pd.DataFrame(pca.fit_transform(df_norm),columns=['PC1','PC2','PC3'])
scores
```

```python
import matplotlib.pyplot as plt
import numpy as np

PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

From the values given above we can identify that :
The first principal component explains 41.4% of the total variation in the dataset.
The second principal component explains 17.9~18% of the total variation.
The third principal component explains 10.9%~ 11% of the total variation.

Scree plot:On the x-axis the principal component is displayed and on the y-axis the percentage of total variance explained by each individual principal component is given.The plot will show a "scree" shape where there is a sharp drop in the magnitude of the eigenvalues(on which scree plot is visualized) after a certain number of components. The point at which the "scree" drops is usually used as an indicator of the number of components to retain.

The scree plot visualizes eigenvalues, the amount of variance each principal component explains. The larger eigenvalue represents the more explained variance. The point of elbow, where the eigenvalue's steep drop from large to small, indicates the appropriate number of principal component to select for the further analysis.

Heatmap helps identify the impact of features on the diagnosis of Benign and Malignant type of tumor.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# create a label encoder
le = LabelEncoder()

# fit and transform the column 'class'
S_df["diagnosis"] = le.fit_transform(S_df["diagnosis"])

# Now you can run KMeans on the modified dataframe
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(S_df.T)

S_df = S_df.sort_values(by='diagnosis')


# k-means clustering on rows
cluster_assignments = kmeans.labels_
S_df=S_df.iloc[:, np.argsort(cluster_assignments)]

# create heatmap
fig = plt.figure(figsize=(25, 20))
sns.heatmap(S_df.corr(),annot=True)
```
